```python
#Exercício 4

lista_clientes = ["Joao","Joana","Jaime","Joaquin","Jeremias"]

def buscar_clientes(cliente, lista_clientes):
    for i in range(len(lista_clientes)):
        if lista_clientes[i] == cliente:
            return i
    return -1

#aviso não colocar else em uma busca de laço for, pois a busca é interrompida no caso de O(n)
```

```python
buscar_clientes("", lista_clientes)
```

```python
pesquisar_cliente = input("Insira o nome do cliente:\n ")
resultado = buscar_clientes(pesquisar_cliente, lista_clientes)
if resultado != -1:
    print(f"Cliente encontrado no índice {resultado}")
else:
    print("Cliente não encontrado")
```

```python
#Exercício 5

def contador_num(n,lista):
    contador = 0
    for i in lista:
        if i == n:
            contador += 1
    return contador
```

```python
contador_num(3,[3,5,6,8,3,6,8,])
```

```python
contador_num(13,[5,6,8,3,6,8,])
```

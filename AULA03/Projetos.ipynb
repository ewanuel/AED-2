```python
#Projeto 1
filmes = [
    "O Poderoso Chefão",
    "Interestelar",
    "Matrix",
    "Titanic",
    "Clube da Luta",
    "O Senhor dos Anéis: A Sociedade do Anel",
    "Pulp Fiction",
    "A Origem",
    "Forrest Gump",
    "Os Bons Companheiros"
];

def buscar_filmes(filme, filmes):
    for i in range(len(filmes)):
        if filmes[i] == filme:
            return print(f"{filme} encontrado na posição: {i}")
    return print (f"filme não encontrado")
        
        
```

```python
buscar_filmes("Pulp Fiction", filmes)
```

```python
buscar_filmes("Pulp ", filmes)
```

```python
#Projeto 2
convidados = ["João", "Maria", "Pedro", "Ana", "Carlos"]


def buscar_convidado(convidado, convidados):
    for i in range(len(convidados)):
        if convidados[i] == convidado:
            return print(f"Essa pessoa é convidada")
    return print (f"Essa pessoa não é convidada")
```

```python
pesquisar = input("digite o nome do convidado")
buscar_convidado(pesquisar, convidados)
```

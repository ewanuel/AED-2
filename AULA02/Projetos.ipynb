```python
#Projeto 1

lista_exalunos = [{"CPF":"123.456.789-01"},
                  {"CPF":"234.567.890-1"},
                  {"CPF":"345.678.901-23"},
                  {"CPF":"456.789.012-34"},
                  {"CPF":"567.890.123-45"}]

dicionario_exaluno = {"123.456.789-01":"Ana Silva", 
                          "234.567.890-12":"Bruno Souza",
                          "345.678.901-23":"Carla Mendes",
                          "456.789.012-34":"Diego Santos",
                          "567.890.123-45":"Elaine Costa" }
def buscar_aluno_linear(lista_exalunos):
    busca_inef = input(f"Digite o CPF do ex-aluno: ")
    for exaluno in lista_exalunos:
        if exaluno["CPF"] == busca_inef:
            return print(f"Ex-aluno encontrado: {busca_inef}")
    return print(f"Ex-aluno não encontrado")


def buscar_aluno_eficiente (dicionario_exaluno):
    busca_ef = input(f"digite o dado do ex-aluno (cpf)\n")
    if busca_ef in dicionario_exaluno:
        return print(f"Ex-aluno encontrado: {dicionario_exaluno[busca_ef]}, Cpf: {busca_ef}")
    else:
        return print(f"Ex-aluno não encontrado")


```

```python
buscar_aluno_eficiente (dicionario_exaluno)
```

```python
buscar_aluno_linear (lista_exalunos)
```

```python
#Projeto 2

hist_cur_al = {
    "João": {
        "disciplinas cursadas": [
            "Cálculo Diferencial e Integral I",
            "Álgebra Linear",
            "Introdução à Programação (Python)",
            "Probabilidade e Estatística",
            "Estruturas de Dados"
        ],
        "disciplinas optativas": [
            "Séries Temporais",
            "Computação em Nuvem para Dados"
        ]
    },
    "Maria": {
        "disciplinas cursadas": [
            "Cálculo Diferencial e Integral I",
            "Álgebra Linear",
            "Introdução à Programação (Python)",
            "Banco de Dados (SQL)",
            "Visualização de Dados"
        ],
        "disciplinas optativas": [
            "Business Intelligence (BI)",
            "Estatística Avançada"
        ]
    },
    "José": {
        "disciplinas cursadas": [
            "Cálculo Diferencial e Integral I",
            "Álgebra Linear",
            "Introdução à Programação (Python)",
            "Aprendizado de Máquina I",
            "Pré-processamento de Dados"
        ],
        "disciplinas optativas": [
            "Processamento de Linguagem Natural (NLP)",
            "Sistemas de Recomendação"
        ]
    },
    "Jaime": {
        "disciplinas cursadas": [
            "Cálculo Diferencial e Integral I",
            "Álgebra Linear",
            "Introdução à Programação (Python)",
            "Banco de Dados (SQL)",
            "Estruturas de Dados"
        ],
        "disciplinas optativas": [
            "Big Data (Spark)",
            "Banco de Dados Não-Relacionais (NoSQL)"
        ]
    },
    "Michel": {
        "disciplinas cursadas": [
            "Cálculo Diferencial e Integral I",
            "Álgebra Linear",
            "Introdução à Programação (Python)",
            "Probabilidade e Estatística",
            "Aprendizado de Máquina I"
        ],
        "disciplinas optativas": [
            "Deep Learning (Redes Neurais)",
            "Visão Computacional"
        ]
    },
    "aluno principal": {
        "disciplinas cursadas": [
            "Probabilidade e Estatística",
            "Aprendizado de Máquina I"
        ],
        "disciplinas optativas": [] 
    }
}

def recomendar_similaridade(hist_cur_al, aluno_principal):
    disc_alvo = set(hist_cur_al[aluno_principal]["disciplinas cursadas"])

    similaridades = {}

    for aluno, dados in hist_cur_al.items():
        if aluno != aluno_principal:
            disc_aluno = set(dados["disciplinas cursadas"])
            similaridade = len(disc_alvo.intersection(disc_aluno))
            similaridades[aluno] = similaridade

    alunos_similares = sorted(similaridades.items(), key=lambda x: x[1], reverse=True)

    recomendacoes = []
    for aluno, _ in alunos_similares[:2]:
        for disciplina in hist_cur_al[aluno]["disciplinas optativas"]:
            if disciplina not in disc_alvo:
                recomendacoes.append({
                    "disciplina": disciplina,
                    "aluno_origem": aluno,
                    "motivo": f"Aluno com {similaridades[aluno]} disciplinas em comum"
                })
    return recomendacoes
    
recomendacoes = recomendar_similaridade(hist_cur_al, "aluno principal")
for i, rec in enumerate(recomendacoes, 1):
    print(f"recomendação de materia:{i}. {rec['disciplina']}")
```

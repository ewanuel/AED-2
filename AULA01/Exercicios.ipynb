```python
#Exercício 2

#conceito de pilha

livros = []

livros.append("O Pequeno Principe ")
livros.append("1984")
livros.append("Dom Quixote")

print(f"Temos uma estante com esses livros: {livros}")

#tal qual uma pilha o ultimo intem adicionado é removido

livro_removido = livros.pop()
print(f"Este livro foi emprestado: {livro_removido}")
print(f"Agora temos uma estante com esses livros: {livros}")
```

```python
#Exercício 3

#conceito de fila

clientes = []

clientes.append("Ana")
clientes.append("Bruno")
clientes.append("Carla")
clientes.append("Daniel")

print(f"temos uma fila com os seguintes clientes {clientes}")

'''pode fazer dessa maneira: cliente_saiu = clientes.remove("Ana")
ou:
'''
cliente_saiu = clientes.pop(0) # o indice 0 corresponde ao primeiro da lista
#tal qual uma fila o primeiro a chegar é o primeiro a ser atendido 
print(f"o(a) cliente: {cliente_saiu} foi atendido(a)")
print(f"temos uma fila com os seguintes clientes {clientes}")

```

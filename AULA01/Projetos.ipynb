```python
#projeto 1

contatos = {}

def adicionar_contato():
    nome = input (f"Favor digitar o nome\n")
    tel = input (f"Favor digitar o telefone\n")
    email = input (f"Favor digitar o email\n")

    contatos[nome] = {"telefone" : tel,"email" : email}

    print(f"Contato {nome} adicionado!")

def buscar_contatos(contatos):
    chave_busca = input (f"Favor digitar o nome que deseja procurar\n")

    if chave_busca in contatos:
        dados = contatos[chave_busca]
        print(f"Nome: {chave_busca}")
        print(f"Telefone: {dados['telefone']}")
        print(f"Email: {dados['email']}")
    else:
        print(f"Contato '{chave_busca}' não encontrado na agenda.")
    
def mostrar_todos_os_contatos(contatos):
    print(f"essa é a lista com todos os contatos: \n {contatos}")
```

```python
adicionar_contato()
```

```python
buscar_contatos(contatos)
```

```python
mostrar_todos_os_contatos(contatos)
```

```python
#projeto 2

lista_de_tarefas = ["Lavar a louça", "Estudar para a prova", "Fazer compras"]

def adicionar_tarefa(lista_de_tarefas):
    tarefa_nova = input (f"Favor digitar a tarefa que o sr(a) gostaria de adicionar \n")

    lista_de_tarefas.append(tarefa_nova)

    print(f"{tarefa_nova} adicionada")

def concluir_tarefa(lista_de_tarefas):
    tarefa_concluida = input (f"Favor digitar a tarefa que foi concluida \n")

    if tarefa_concluida in lista_de_tarefas:
        lista_de_tarefas.remove(tarefa_concluida)
        print(f"{tarefa_concluida} removida")
    else:
        print(f"tarefa não encontrada")

def mostrar_todas_as_tarefas(lista_de_tarefas):
    print(f"essa é a lista com todas as tarefas: \n {lista_de_tarefas}")
```

```python
adicionar_tarefa(lista_de_tarefas)
```

```python
concluir_tarefa(lista_de_tarefas)
```

```python
mostrar_todas_as_tarefas(lista_de_tarefas)
```
